```python
import cv2
from PIL import Image
```

```python
img = cv2.imread('/content/drive/Othercomputers/My Mac/med-proj/backend/notebooks/dark_image.jpg', flags=cv2.IMREAD_GRAYSCALE)
img
```

```python
_, new_img = cv2.threshold(img, 150, 255, cv2.THRESH_BINARY) #threshold -> set to 150 / tuple is returned -> ignore 1st value
Image.fromarray(new_img)
```

```python
new_image = cv2.adaptiveThreshold(
    img, 255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C, #some way to divide image and get specific threshold per region
    cv2.THRESH_BINARY, #needed
    61, #block size
    11 #C value
) #need to play around w values
Image.fromarray(new_image)
```
