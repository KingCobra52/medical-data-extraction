```python
import re
```

Extract all numbers from text

```python
text = "Patient's phone is 7321119999. Bill amount is 120$"
pattern = '\d+'
match = re.findall(pattern,text)
match
```

Extract only phone number. Phone number is always 10 digits.


```python
text = "Patient's phone is 7321119999. Bill amount is 120$"
pattern = '\d{10}'
match = re.findall(pattern,text)
match
```

Extract phone number, which is either 10 continuous digits or follows this format (xxx)-xxx-xxxx

```python
text = "Patient's phone is (732)-111-9999, spouse phone number 7326664444. Bill amount is 120$"
pattern = '\(\d{3}\)-\d{3}-\d{4}|\d{10}'
match = re.findall(pattern, text)
match
```

Extract phone number and bill amount seperately

```python
text = "Patient's phone is 7321119999. Bill amount is 120$"
pattern = '(\d{10})\D+(\d+)\$'
match = re.search(pattern, text)
match
```

```python
phone_number, bill_amount = match.groups() #groups -> whatever is in the brackets
```

```python
phone_number
```

```python
bill_amount
```

```python
text = '''Name: Marta Sharapova Date: 5/11/2022

Address: 9 tennis court, new Russia, DC

K



Prednisone 20 mg
Lialda 2.4 gram

Directions:

Prednisone, Taper 5 mg every 3 days,
Finish in 2.5 weeks a
Lialda - take 2 pill everyday for 1 month

Refill: 2 times '''
```

```python
pattern = "Name(.*)Date"
match = re.findall(pattern, text)
match[0].strip()
```

```python
pattern = 'Address:(.*)\n'
match = re.findall(pattern, text)
match[0].strip()
```

```python
pattern = 'Address:[^\n]*(.*)Directions'
match = re.findall(pattern, text, flags=re.DOTALL) #flags parameter creates inclusion of \n
print(match[0].strip())
```

```python
pattern = 'Directions:(.*)Refill'
match = re.findall(pattern, text, flags=re.DOTALL)
print(match[0].strip())
```

```python
pattern = 'Refill:(.*)times'
match = re.findall(pattern, text, flags=re.DOTALL)
match[0].strip()
```
